```python
from PIL import Image, ImageDraw, ImageFont
```

```python
def make_image(text, background_color=(255,255,255), text_color=(0,0,0)):
    fontsize = 300
    w = 400
    line_thickness = 6
    scale = 4
    h = w

    image = Image.new("RGBA", (w*scale, h*scale), background_color)
    draw = ImageDraw.Draw(image)
    font = ImageFont.truetype("B612 Mono-Regular.ttf", fontsize)

    text_width, text_height = font.getsize(text)
    draw.text((scale*w/2 - text_width/2, scale*h/2-int(1.4*text_height/2)), text, text_color, font=font)

    dx = int(0.1*w)
    dy = int(0.15*h)
    x0 = scale*dx
    y0 = scale*dy
    x1 = scale*w-x0
    y1 = scale*h-y0
    draw.line(((x0, y0), (x1, y0)), fill=text_color, width=scale*line_thickness)
    draw.line(((x0, y1), (x1, y1)), fill=text_color, width=scale*line_thickness)

    img_resized = image.resize((w, h), Image.ANTIALIAS)
    
    return img_resized
```

```python
make_image('CENTER', background_color=(200, 200, 200))
```

```python
for name in "LEFT CENTER RIGHT".split(" "):
    fn = "../examples/text_{}.png".format(name)
    img = make_image(name, background_color=(200, 200, 200))
    img.save(fn)
```
